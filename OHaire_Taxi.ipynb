```python
import gym
import matplotlib.pyplot as plt
import numpy as np
import os
import time


taxi= gym.make('Taxi-v3')
board=taxi.env
board.reset()
board.s = board.encode(3, 1, 2, 0) #Sets location of car and passanger

gamma=0.9
observation_space = board.observation_space.n
action_space = board.action_space.n

print("Action Space {}".format(action_space))
print("State Space {}".format(observation_space))
print("")
print("____Initial Board____")
board.render()

policy = np.random.randint(0,action_space,(observation_space)) #Fills in random actions into policy space

V_s_dot = np.zeros(observation_space)
r = np.zeros((observation_space, action_space, observation_space))
p = np.zeros((observation_space, action_space, observation_space))


for curr_action in range(action_space):
    for curr_state in range(observation_space):
        prob=board.P[curr_state][curr_action][0][0]
        next=board.P[curr_state][curr_action][0][1]
        reward=board.P[curr_state][curr_action][0][2]
        end=board.P[curr_state][curr_action][0][3]
        r[curr_state, curr_action, next] = reward #Fillls r vector
        p[curr_state, curr_action, next] = prob #Fills p vector

finished=False
count=0
while not finished:
    previous_policy = policy.copy()
    previous_V_s_dot = V_s_dot.copy()
    V_s = np.einsum('ijk,ijk -> ij', p, r + gamma * V_s_dot)
    reshaped_policy = np.zeros((observation_space, action_space))
    reshaped_policy[np.arange(observation_space), policy] = 1
    V_s_dot = np.sum(reshaped_policy * V_s, 1)
    policy = np.argmax(V_s, axis=1)
    count=count+1
    if (policy==previous_policy).all(): #If the policy is equal to the previous policy, stop loop
        finished=True

done = False
action = policy[board.s]
while not done:
    state, reward, done, info = board.step(action)
    action = policy[state] #Changes policy to actions
    board.render()
print ("Completed using "+str(count)+" iterations of Policy Iteration!")
```
